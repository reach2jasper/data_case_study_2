```python
import pandas as pd
import numpy as np

import pymysql                        # for getting data from a SQL database
from sqlalchemy import create_engine  # for establishing the connection and authentication

from getpass import getpass  # To get the password without showing the input
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

```python
#1. How many distinct (different) actors' last names are there?
data = pd.read_sql_query('SELECT last_name, count(distinct(last_name)) as count FROM sakila.actor GROUP BY last_name;', engine)
data.head()
```

```python
#2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
data1 = pd.read_sql_query("""SELECT *, CASE WHEN DAYOFWEEK(rental_date) 
IN (1,7) 
THEN 'weekend' 
ELSE 'workday' 
END AS day_type FROM rental;""", engine)
data1
```

```python
#3. Get all films with ARMAGEDDON in the title.
data3 = pd.read_sql_query("SELECT title FROM sakila.film WHERE title LIKE '%%ARMAGEDDON%%';",engine)
data3
```

```python
#4. Get 10 the longest film
data4 = pd.read_sql_query("SELECT title, length FROM sakila.film ORDER BY length DESC LIMIT 10;", engine)
data4
```

```python
#5. How many films include Behind the Scenes content?
data5 = pd.read_sql_query("""SELECT count(*) as count FROM film 
WHERE special_features 
LIKE 'Behind the Scenes';""", engine)
data5
```

```python
#6. Which kind of movies (rating) have a mean duration of more than two hours? I assumed to display the ratings 
data6 = pd.read_sql_query("SELECT rating, count(rating) FROM sakila.film WHERE length>120 GROUP BY rating;",engine)
data6
```

```python
#7. # Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.
data7 = pd.read_sql_query("""SELECT title, length, RANK() OVER(ORDER BY length DESC) AS rank_film 
FROM sakila.film 
WHERE length IS NOT NULL;""", engine)
data7
```
